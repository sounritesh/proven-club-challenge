```python
from src.dataset import NewsDataset
```

```python
train_dataset = NewsDataset(train_X, train_y, scaler)
test_dataset = NewsDataset(test_X, test_y, scaler)
```

```python
train_dataset[0]
```

```python
from torch.utils.data import DataLoader

train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True, drop_last=True)
test_loader = DataLoader(test_dataset, batch_size=8, shuffle=True, drop_last=True)
```

```python
import torch.nn as nn

class NewsModel(nn.Module):
    def __init__(self, num_features = 51, batch_size = 8):
        super().__init__()
        self.flatten = nn.Flatten()
        self.linear_stack = nn.Sequential(
            nn.Linear(batch_size * num_features, 256),
            nn.ReLU(),
            nn.Linear(256, 256),
            nn.ReLU(),
            nn.Linear(256, batch_size)
        )
        
    def forward(self, X):
        x = self.flatten(X).reshape(-1)
        return self.linear_stack(x)
```

```python
model = NewsModel()

device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)

print(device, "\n", model)
```

Training model for 10 epochs

```python
list(model.parameters())
```

```python
## Setting params
from torch.optim import Adam
from torch.nn import MSELoss

lr = 1e-3
epochs = 10
optimizer = Adam(list(model.parameters()), lr)
loss_fn = MSELoss()
```

```python
!pip install tqdm
!pip install ipywidgets
```

```python
from tqdm.notebook import tqdm

for epoch in range(epochs):
    optimizer.zero_grad()
    loss_total = 0
    for data in train_loader:
        features = data["features"]
        targets = data["target"]
        
        features = features.to(device)
        
        results = model(features)
        loss = loss_fn(results.to("cpu"), targets)
        
        loss_total += loss.item()
        
        loss.backward()
        optimizer.step()
    print("{}: {}".format(epoch+1, loss_total/len(train_loader)))
```

```python
from sklearn.metrics import r2_score
```

```python
res = []
model.eval()
with torch.no_grad():
    for data in test_loader:
        features = data["features"]
        targets = data["target"]
        
        features = features.to(device)
        
        results = model(features)
        
        res.extend(results.reshape(-1).to("cpu"))
```

```python
nn_results = r2_score(test_y[:len(res)], res)
```

```python
print(abs(nn_results))
```

##### Final Results (r2 scores)

```python
print("Random Forest: {}, Linear Regresion: {}, Neural Network: {}".format(abs(rf_results), abs(lr_results), abs(nn_results)))
```

As observed above, Random Forest is our best performing model on the unseen test dataset wherease the Linear Regression classifier is the worst. Note it is hard to quantify these results due to not very correlated features to targets.
