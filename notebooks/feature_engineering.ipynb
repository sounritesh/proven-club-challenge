##### Understanding other features

Understanding correaltion of article features and popularity (number of shares)

```python
corr_df = df.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(corr_df)

plt.figure(figsize=(20, 20))
plt.show()
```

```python
for i, row in corr_df.iterrows():
    if i != " shares":
        if row[" shares"] >= 0.1 or row[" shares"] <= -0.1:
            print("{}: {}".format(i, row[" shares"]))
```

According to the preliminary correlation analysis, we discover that none of the features in the dataset have a high positive or negative correlation with popularity.

Let us look at the inter feature correlation to figure out if we can omit certain features from the dataset on account of them being highly correlated

```python
correlated_cols = []

for i, row in corr_df.iterrows():
    for col in cols:
        if (row[col] >= 0.75 or row[col] <= -0.75) and i != col:
            _t = (i, col)
            t_ = (col, i)
            if _t not in correlated_cols and t_ not in correlated_cols:
                correlated_cols.append((i, col))
                print("({}, {}): {}".format(i, col, row[col]))
```

```python
len(correlated_cols)
```

There are 11 column pairs with very high positive or negative correlation. Since, these 11 pairs are so highly correlated, it doesn't make sense to keep both cols of a pair. Hence, we pick one column from each pair.

```python
shortlisted_related_cols = []
for cols in correlated_cols:
    if cols[0] in shortlisted_related_cols or cols[1] in shortlisted_related_cols:
        continue
    
    shortlisted_related_cols.append(cols[0])
```

```python
shortlisted_related_cols
```

```python
cols_to_remove = []
for cols in correlated_cols:
    if cols[0] not in shortlisted_related_cols:
        cols_to_remove.append(cols[0])
        
    if cols[1] not in shortlisted_related_cols:
        cols_to_remove.append(cols[1])
        
cols_to_remove = set(cols_to_remove)
```

```python
cols_to_remove
```

Removing the above 8 columns from our final dataset

```python
df = df[list(set(df.columns) - cols_to_remove)]
```
